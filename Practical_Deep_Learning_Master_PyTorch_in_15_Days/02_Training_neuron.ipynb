```python
import torch
from torch import nn

# Input: Temperature in 째C
X1 = torch.tensor([[10]], dtype=torch.float32)
# Actual value: Temperature 째F
y1 = torch.tensor([[50]], dtype=torch.float32)

# Input: Temperature in 째C
X2 = torch.tensor([[37.78]], dtype=torch.float32)
# Actual value: Temperature 째F
y2 = torch.tensor([[100.0]], dtype=torch.float32)
```

### 1st training

```python
model = nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

print("\n======Before training======\n")
print(model.bias)
print(model.weight)
print(model.parameters)
# Training pass
optimizer.zero_grad()
outputs = model(X1)
loss = loss_fn(outputs, y1)
loss.backward()
optimizer.step()

print("\n======After training======\n")
print(model.bias)
print(model.weight)
print(model.parameters)

print("\n\nPrediction")
y1_pred = model(X1)
print("y1_pred =", y1_pred)
```

### 2nd training

```python
model = nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001) # If you're getting prediction as nan. Then decrease the learning rate more..

for i in range(0, 100000): # change from 10000 -> 50000 -> 100000 to see the prediction difference
    # Training pass
    optimizer.zero_grad()
    outputs = model(X1)
    loss = loss_fn(outputs, y1)
    loss.backward()
    optimizer.step()

    optimizer.zero_grad()
    outputs = model(X2)
    loss = loss_fn(outputs, y2)
    loss.backward()
    optimizer.step()

    if i % 100 == 0:
        print(model.bias)
        print(model.weight)

y1_pred = model(X1)
print("y1_pred =", y1_pred)
```

### Batch learning

```python
import torch
from torch import nn

# Input: Temperature in 째C
X = torch.tensor([
    [10],
    [37.78]
], dtype=torch.float32)

# Actual value: Temperature 째F
y = torch.tensor([
    [50],
    [100.0]
], dtype=torch.float32)
```

```python
model = nn.Linear(1, 1)
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.0001)

for i in range(0, 150000):
    # Training pass
    optimizer.zero_grad()
    outputs = model(X)
    loss = loss_fn(outputs, y)
    loss.backward()
    optimizer.step()

    if i % 100 == 0:
        print(model.bias)
        print(model.weight)

print("----")

measurements = torch.tensor([
    [37.5]
], dtype=torch.float32)

model.eval()
with torch.no_grad():
    prediction = model(measurements)
    print(prediction)
```

37.5 Celcius is approximately 99.4993(~100) Fahrenheit
